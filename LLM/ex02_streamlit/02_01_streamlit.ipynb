## Streamlit이란?

**Streamlit**은 **파이썬만으로 웹 앱을 쉽게 만들 수 있게 해주는 오픈소스 프레임워크**입니다.

- 특히 **데이터 분석, 시각화, 머신러닝 결과 공유**에 많이 사용됩니다.
- 복잡한 HTML, CSS, JavaScript를 직접 작성하지 않아도 파이썬 코드 몇 줄로 **인터랙티브한 웹 화면**을 만들 수 있습니다.

---

## Streamlit 장점

- **파이썬만으로 앱 제작 가능**

- **차트, 표, 텍스트, 이미지 등을 쉽게 출력 가능**

- **버튼, 슬라이더, 입력창 같은 위젯 제공**

- **머신러닝 모델 데모 화면 제작에 적합**

- **빠르게 프로토타입 제작 가능**

- 데이터 과학자·AI 엔지니어·분석가 들이 웹 프론트엔드 지식 없이도 파이썬 코드 몇 줄로 대화형 앱을 만들 수 있다는 점이 가장 큰 장점

---
