## 1. SQLite란?

SQLite는 **파일 기반의 관계형 데이터베이스(RDBMS)** 입니다.

- 데이터베이스(Database)란 여러 데이터를 체계적으로 저장하고 관리하기 위한 저장소입니다.

    - 단순한 파일 저장이 아니라 검색, 수정, 삭제 등을 효율적으로 할 수 있도록 구조화된 데이터 집합을 의미합니다.

        | id | name | age |
        | :--- | :--- | :--- |
        | 1 | Kim | 25 |
        | 2 | Lee | 30 |

    - 위와 같은 데이터가 테이블 형태로 저장되어 관리됩니다.

    - 데이터베이스의 주요 기능

        | 기능 | 설명 |
        | :--- | :--- |
        | 저장 (Storage) | 데이터를 안전하게 저장 |
        | 검색 (Query) | 원하는 데이터를 빠르게 조회 |
        | 수정 (Update) | 데이터 변경 |
        | 삭제 (Delete) | 필요없는 데이터 제거 |
        | 관리 (Management) | 데이터 일관성 유지 |

    - 대표적인 데이터베이스

        - MySQL, PostgreSQL, Oracle, SQLite

- 2. RDBMS (Relational Database Management System) : 관계형 데이터베이스를 관리하는 시스템입니다.

    - 데이터를 테이블(Table) 형태로 저장하고 테이블 간의 관계(relation) 를 이용해 데이터를 관리합니다.

    - users 테이블

        | id | name | age |
        | :--- | :--- | :--- |
        | 1 | Kim | 25 |
        | 2 | Lee | 30 |
    
    - orders 테이블

        | id | user_id | product |
        | :--- | :--- | :--- |
        | 1 | 1 | laptop |
        | 2 | 2 | phone |

    - orders 테이블의 user_id가 users 테이블의 id를 참조하여 두 테이블이 연결됩니다.  
    테이블 간 관계(Relation) 가 만들어집니다.

### 특징

- 별도의 **데이터베이스 서버가 필요 없음**
- 하나의 **파일(.db)** 로 데이터베이스 관리
- Python에 **기본 내장 라이브러리(sqlite3)** 존재
- 가볍고 빠르기 때문에 **실습, 모바일 앱, 소형 서비스**에서 많이 사용됩니다.

예

```txt

↓
example.db (파일)
```

## 3. 데이터베이스 생성 및 연결

SQLite에서는 하나의 파일이 데이터베이스입니다.

````python
import sqlite3

# example.db 데이터베이스에 연결
# 파일이 없으면 자동으로 생성됩니다.
conn = sqlite3.connect("example.db")

print("데이터베이스 연결 완료")
````

실행 후 폴더에 다음 파일이 생성됩니다.

## 4. Cursor 생성

- Cursor 는 데이터베이스에 SQL 명령을 실행하고 결과를 가져오는 객체입니다.

- 데이터베이스와의 연결(Connection)은 접속만 담당하며, 실제 SQL 실행은 Cursor 객체가 처리합니다.

    즉, 데이터 베이스 작업은 다음과 같은 구조로 이루어집니다.

    ```txt
    Python 프로그램
          ↓
    Connection (데이터베이스 연결)
          ↓
    Cursor (SQL 실행)
          ↓
    SQLite 데이터베이스
    ```

- 따라서 SQL을 실행하려면 먼저 Cursor를 생성해야 합니다.

````python
cursor = conn.cursor()
````

 ---
## 5. 테이블 생성 (CREATE TABLE)

- cursor.execute()는 SQL 문장을 데이터베이스에 전달하여 실행하는 함수입니다.

````python
cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER
)
""")

# commit() 은 데이터베이스 변경 내용을 실제 파일에 저장하는 명령입니다.
conn.commit()

print("테이블 생성 완료")
````

- 테이블 구조

```txt
users
-------------------------
id      INTEGER
name    TEXT
age     INTEGER
```

 ---

## 6. 데이터 삽입 (INSERT)

- INSERT INTO 는 테이블에 새로운 데이터를 추가하는 SQL 명령입니다.

- "INSERT INTO users (name, age) VALUES (?, ?)", ("Kim", 25)

    - users 테이블에 Kim, 25 데이터를 추가합니다.
    - ? 는 값이 들어갈 자리 표시자(placeholder) 입니다.
    - ("Kim", 25)
        | ? 위치 | 값 |
        | :--- | :--- |
        | 첫 번째 ? | "Kim" |
        | 두 번째 ? | 25 |

````python
cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Kim", 25))
cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Lee", 30))


# commit() 은 데이터베이스 변경 내용을 실제 DB 파일에 저장하는 명령입니다.
conn.commit()

print("데이터 저장 완료")
````

```txt
테이블 데이터

users
-------------------------
id | name | age
1  | Kim  | 25
2  | Lee  | 30
```

## 7. 데이터 조회 (SELECT)

- SELECT * FROM users : users 테이블의 모든 컬럼을 조회하라는 의미입니다.

- cursor.fetchall() : 조회 결과를 모두 가져온느 함수입니다. 조회 결과는 리스트(list) 형태로 반환됩니다.

````python
cursor.execute("SELECT * FROM users")

rows = cursor.fetchall()

for row in rows:
    print(row)
````

각 데이터는 튜플(tuple) 형태입니다.

---

## 8. 데이터 수정 (UPDATE)

- users 테이블에서 특정 사용자의 나이를 수정하는 코드입니다.

````python
cursor.execute(
    "UPDATE users SET age = ? WHERE name = ?",
    (26, "Kim")
)

conn.commit()

print("데이터 수정 완료")
````

````python
cursor.execute("SELECT * FROM users")

rows = cursor.fetchall()

for row in rows:
    print(row)
````

 ---

## 9. 데이터 삭제 (DELETE)

- users 테이블에서 특정 데이터를 삭제하는 코드입니다.

- (count, ) : 이 값은 SQL의 ? 자리에 들어간느 파라미터 튜플(tuple) 입니다.

    - 쉼표 , 가 있는 이유는 Python에서 값이 1개인 튜플을 만들기 위해서입니다.

````python
cursor.execute("DELETE FROM users WHERE name = ?", ("Lee",))

conn.commit()

print("데이터 삭제 완료")
````

````python
cursor.execute("SELECT * FROM users")

rows = cursor.fetchall()

for row in rows:
    print(row)
````
