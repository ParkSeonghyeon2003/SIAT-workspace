# Audio and Speech

- **Speech** : 사람이 말하는 음성
- **Audio** : 음성, 음악, 효과음 등 **모든 소리 데이터를 포함하는 개념**

즉, Speech ⊂ Audio

Speech는 Audio의 한 종류입니다.

# OpenAI 오디오 사용 사례

오디오는 **사람의 음성 뿐 아니라 음악, 효과음 등 다양한 소리 데이터**를 포함합니다.

OpenAI는 다음과 같은 작업을 수행할 수 있는 다양한 API를 제공합니다.
- 오디오 입력 처리
- 오디오 생성
- 오디오와 텍스트를 동시에 처리

 ---

# 음성 에이전트 (Voice Agents)

음성 에이전트는 **사용자의 음성을 이해하고 작업을 수행한 뒤 음성으로 응답하는 AI 시스템**입니다.

예
- 음성 비서
- 음성 챗봇
- 실시간 상담 시스템

--- 

## 1. Speech-to-Speech 방식

- **Speech-to-Speech 모델 + Realtime API 사용**

특징
- 입력 음성을 바로 음성으로 응답
- 지연 시간이 매우 짧음 (low latency)
- 자연스러운 대화 가능
- 음성 → AI 처리 → 음성

---

## 2. 체이닝 방식 (Chaining 방식, 기존 방식)

예전에는 다음과 같이 여러 모델을 **순차적으로 연결**하여 사용했습니다.

음성(STT)
↓
텍스트(LLM)
↓
음성(TTS)

예

사용자 : "무지개에서 두 번째 색은 뭐야?"

1. STT → 텍스트 변환
2. LLM → 질문 처리
3. TTS → 음성 응답

특징
- 기존 텍스트 기반 에이전트를 쉽게 확장 가능
- 안정성과 제어성이 높음

# OpenAI Audio API 특징 및 선택 기준

| API | 지원 기능 | 스트리밍 지원 |
| :--- | :--- | :--- |
| Realtime API | 오디오 및 텍스트 입력/출력 | 오디오 입력 및 출력 스트리밍 |
| Chat Completions API | 오디오 및 텍스트 입력/출력 | 오디오 출력 스트리밍 |
| Transcription API | 오디오 입력 → 텍스트 출력 | 텍스트 스트리밍 |
| Speech API | 텍스트 입력 → 오디오 출력 | 오디오 스트리밍 |

사용 목적에 따른 선택

- **양방향 실시간 음성 대화**  
→ Realtime API

- **텍스트를 음성으로 변환**  
→ Speech API 또는 Chat Completions API

- **음성을 텍스트로 변환**  
→ Transcription API

### whisper-1

- OpenAI에서 개발한 자동 음성 인식(ASR, Automatic Speech Recognition) 모델
- 68만 시간 이상의 다국어 오디오 데이터로 학습된 Whisper 모델 기반
- OpenAI API를 통해 음성 → 텍스트 변환(STT)에 사용

주요 기능

- 음성 → 텍스트 변환 (transcription)
- 음성 → 영어 번역 (translation)

장점

- 다국어 지원 (한국어 포함 수십 개 언어)
- 높은 정확도
- 잡음 환경에서도 비교적 강건함
- 긴 오디오 처리 가능

활용 분야

- 자막 생성
- 회의록 자동 생성
- 팟캐스트 전사
- 음성 데이터 분석
- 음성 기반 검색 시스템

OpenAI STT 모델 종류

- gpt-4o-transcibe
- gpt-4o-mini-transcribe
- whisper-1
- gpt-4o-transcribe-diarize (화자 분리)
