### 펑션 콜링(Function Calling)

#### 1. 평션 콜링이란

펑션 콜링(Function Calling)은 거대 언어 모델(LLM)이 단순히 텍스트를 생성하는 것을 넘어서 **외부 함수(코드, API)를 안전하고 구조적으로 호출할 수 있도록 하는 메커니즘**입니다.

- 기존 LLM

    `사용자 질문 → 자연어 답변 생성`

- 평션 콜링 사용

    `사용자 질문 → LLM이 적절한 함수 선택 → 함수 실행 → 결과 기반 답변 생성`

즉, LLM이 **외부 프로그램과 협력하여 작업을 수행**할 수 있게 됩니다.

 ---
#### 2. 기존 LLM 방식

기존에는 LLM이 **자연어 답변만 생성**했습니다.

예)

사용자 : 지역의 맛집 알려줘  
LLM : 이 지역에는 유명한 맛집이 여러 곳 있습니다. 한식당, 분식집, 카페 등이 인기 있습니다.

문제점

- 실제 지역 정보를 조회하지 않음

- 현재 위치를 반영하지 않을 수 있음

- 최신 정보가 아닐 수 있음

- 영업 종료, 이전, 폐업 여부를 알 수 없음

- 정확성이 떨어질 수 있음

 ---
#### 3. 펑션 콜링 방식

펑션 콜링을 사용하면 LLM이 **답변 대신 함수 호출 정보를 반환**합니다.

예)

- 사용자 질문

    `오늘 애플 주가 알려줘`

- LLM 응답

    ```json
    {
        "name": "get_stock_price",
        "argumentst": {
            "ticker": "AAPL"
        }
    }
    ```

- 애플리케이션 동작

    ```
    1. LLM이 함수 호출 정보 반환
    2. 프로그램이 실제 함수 실행
    3. 결과를 다시 LLM에 전달
    4. LLM이 최종 자연어 답변 생성
    ```
