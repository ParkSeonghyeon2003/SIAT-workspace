# PyMuPDF (fitz) 가이드

**PyMuPDF**는 Python에서 PDF, XPS, EPUB 등 다양한 문서 형식을 처리하기 위한 강력한 라이브러리입니다. 내부적을 고성능 문서 렌더링 엔진인 **MuPDF**를 사용하여 타 라이브러리 대비 압도적인 속도와 정확도를 자랑합니다.

PyMuPDF라는 이름은 이 라이브러리의 정체성을 보여주는 세 가지 요소가 결합된 약어입니다.

- Py: Python을 의미합니다. 파이썬 환경에서 사용할 수 있도록 바인딩되었다는 뜻입니다.
- Mu: 내부 핵심 엔진인 MuPDF의 'Mu'를 가져온 것입니다.  
    참고) MuPDF에서 'Mu'($\mu$)는 그리스 문자로, 소프트웨어 공학에서 보통 '작고 가볍다(Micro)'는 의미로 쓰입니다.
- PDF: 가장 주된 처리 대상인 Portable Document Format을 의미합니다.

---

## 주요 특징

- **초고속 성능:** C로 작성된 MuPDF 엔진을 기반으로 하여 처리 속도가 매우 빠릅니다.
- **높은 정확도:** 텍스트 레이아웃 유지 및 폰트 렌더링 성능이 뛰어납니다.
- **다양한 활용도:** 단순 텍스트 추출을 넘어 이미지 변환, 주석 추가, 문서 병합 등이 가능합니다.

---

## 주요 기능

1. **텍스트 추출:** PDF 내의 텍스트를 구조화된 형태나 단순 문자열로 추출
2. **페이지 변환:** PDF 페이지를 PNG, JPG 등 이미지 파일로 렌더링
3. **이미지 관리:** PDF 내부에 삽입된 이미지 객체 추출 및 교체
4. **문서 편집:** 텍스트/주석(Annotation) 추가, 페이지 삭제 및 결합
5. **표(Table) 추출:** 문서 내 표 데이터를 감지하고 구조화된 데이터로 변환
6. **전처리 도구:** RAG(검색 증강 생성), 문서 검색 시스템, 데이터 분석용 데이터 정제

텍스트 추출하기

```python
# PyMuPDF 라이브러리를 fitz라는 이름으로 불러옵니다.
# PDF 문서를 열고, 페이지를 읽고, 텍스트나 이미지를 추출할 때 사용합니다.
import fitz


# fitz.open() : PDF 파일을 열어 Document 객체로 생성합니다.
# "data/crop-model.pdf" 파일을 읽어 메모리에 로드합니다.
with fitz.open("data/crop-model.pdf") as doc:


    # PDF 문서의 모든 페이지를 하나씩 반복합니다.
    # doc은 페이지들의 집합(Document 객체)이므로 for문으로 순회할 수 있습니다.
    for number, page in enumerate(doc):

        # page.get_text() : 햔제 페이지에 있는 텍스트를 추출합니다.
        # PDF 내부에 저장된 텍스트 데이터를 문자열 형태로 반환합니다.
        text = page.get_text()

        # 추출된 텍스트를 화면에 출력합니다.
        # 각 페이지의 텍스트가 순서대로 콘솔에 표시됩니다.
        print(f"{number + 1}페이지")
        print(text[:50])
        print()
```

 ---
고해상도 이미지 생성

```python
# PyMuPDF 라이브러리를 fitz라는 이름으로 불러옵니다.
# PyMuPDF는 PDF 문서를 읽고, 수정하고, 이미지로 변환하는 긴으을 제공합니다.
import fitz


with fitz.open("data/crop-model.pdf") as doc:


    # PDF 문서의 첫 번째 페이지를 가져옵니다.
    # PDF 페이지는 리스트처럼 접근할 수 있으며 doc[0] 은 첫 번째 페이지를 의미합니다.
    page = doc[0]

    
    # 확대 비율 설정
    # 기본 해상도는 1이며, 2로 설정하면 2배 해상도로 렌더링됩니다.
    # 값이 클수록 더 선명한 이미지를 만들 수 있습니다.
    zoom = 2


    # 변환 행렬(Matrix) 생성
    # Matrix(가로 확대, 세로 확대)
    # 여기서는 가로와 세로 모두 2배 확대합니다.
    matrix = fitz.Matrix(zoom, zoom)


    # 페이지를 이미지(Pixmap)로 변환합니다.
    # matrix 옵션을 사용하면 확대된 해상도로 이미지를 생성할 수 있습니다.
    pix = page.get_pixmap(matrix=matrix)


    # 생성된 Pixmap 이미지를 파일로 저장합니다.
    pix.save("output/page_high_res.png")
```

PDF 내부 이미지 추출

```python
# PyMuPDF 라이브러리를 fitz라는 이름으로 불러옵니다.
# PDF 문서를 열고 텍스트나 이미지를 추출할 때 사용하는 라이브러리입니다.
import fitz


# 폴더 생성 여부 확인을 위해 os 모듈을 불러옵니다.

# 저장할 이미지 폴더 이름
output_dir = "output/crop/images"

# os.makedirs()
# 지정한 폴더가 없으면 생성합니다.
# exist_ok=True : 폴더가 이미 존재해도 에러가 발생하지 않습니다.
os.makedirs(output_dir, exist_ok=True)


# fitz.open() : PDF 파일을 열어 Document 객체로 생성합니다.
# "crop-model.pdf" 파일을 메모리로 로드합니다.
with fitz.open("data/crop-model.pdf") as doc:

    page = doc[0]

    # page.get_images() :
    # 현재 페이지에 포함된 모든 이미지 정보를 리스트 형태로 가져옵니다.
    # 각 이미지 정보는 튜플 형태로 저장됩니다.
    # [(xref, smask, width, height, bpc, colorspace, ...)]
    images = page.get_images()

    # 페이지 안에 있는 이미지를 하나씩 반복 처리합니다.
    for i, img in enumerate(images):
        # (xref, smask, width, height, bpc, colorspace, ...)
        # img[0] : 이미지의 xref 번호
        # xref는 PDF 내부에서 이미지 객체를 식별하는 고유 번호입니다.
        xref = img[0]

        # doc.extract_image(xref)
        # xref 번호를 이용해 실제 이미지 데이터를 추출합니다.
        base_image = doc.extract_image(xref)

        # 추출된 이미지 데이터 중 실제 이미지 바이트를 가져옵니다.
        # "image" 키에는 이미지의 바이너리 데이터가 저장되어 있습니다.
        image_bytes = base_image["image"]

        # 이미지 파일을 바이너리 쓰기 모드("wb")로 생성합니다.
        # "output/crop/images_0.png", "output/crop/images_1.png" 만들어 집니다.
        with open(f"{output_dir}/image_{i}.png", "wb") as f:
            f.write(image_bytes)
```

```python
# PyMuPDF 라이브러리
import fitz

# table.to_pandas()에서 내부적으로 pandas DataFrame 객체를 생성합니다.
# pandas가 설치되어 있어야 동작합니다.
import pandas as pd

# 출력 폴더 생성
import os
os.makedirs("output", exist_ok=True)


# PDF 파일 열기
with fitz.open("data/extracted_tables.pdf") as doc:

    # 모든 페이지 반복
    for page_num, page in enumerate(doc):

        # 페이지에서 표를 찾습니다.
        tabs = page.find_tables()

        # 표가 없으면 다음 페이지로
        if len(tabs.tables) == 0:
            continue

        print(f"\n[페이지 {page_num+1}] 표 {len(tabs.tables)}개 발견")

        # 여러 표가 있을 수 있으므로 반복
        for i, table in enumerate(tabs.tables):

            # 표를 pandas DataFrame으로 변환
            df = table.to_pandas()

            print(df)

            df.to_excel(
                f"output/page_{page_num+1}_table_{i+1}.xlsx",
                index=False
            )
```

### 스캔 PDF vs. 일반 PDF

- 일반 PDF (Digital Native): 워드, 한글, 파워포인트 등에서 **"PDF로 저장"** 기능을 이용해 만든 파일입니다.  
문서 내부에 **텍스트 정보가 그대로 저장되어 있습니다.**

### 2. 스캔 PDF (Image-based)

종이 문서를 스캐너로 읽거나 사진을 찍어서 만든 PDF입니다.

- PDF 내부에 **텍스트 정보가 없는 경우가 많습니다.**
- 문서 전체가 이미지처럼 들어 있습니다.
- 컴퓨터는 글자를 글자로 보는 것이 아니라  
**검은 점, 흰 점, 색 점들의 모임(픽셀)** 로 인식합니다.

즉, 컴퓨터 입장에서는:

"이건 글자가 아니라 그냥 이미지다. 무엇이 적혀 있는지 바로는 알 수 없습니다."

그래서 이런 특징이 있습니다.

- 복사 어려움

- 검색 불가

- `page.get_text()` 결과가 거의 없거나 비어 있음

- OCR(광학 문자 인식)이 필요

- 스캔 PDF vs. 일반 PDF 확인 방법

    - 가장 간단한 방법 : **PDF 위에서 마우스로 글자를 드래그 해보기**

        | 결과 | PDF 종류 |
        | :--- | :--- |
        | 글자가 선택됨 | 일반 PDF |
        | 선택이 안 되고 전체 이미지처럼 움직임 | 스캔 PDF |

OCR이란?

- OCR(Optical Character Recognition)은 우리말로 '광학 문자 인식'이라고 합니다.

- 이미지 안에 있는 기하학적인 모양(선, 점, 곡선)을 분석해서 이미지를 다시 텍스트 데이터로 변환해주는 기술입니다.

PyMuPDF 자체만으로는 이미지를 글자로 읽어내는 기능(OCR)이 없습니다. 그래서 보통 다음과 같은 OCR 엔진을 함께 사용합니다.

- Tesseract: 가장 유명한 **오픈소스 OCR 엔진**, Google이 개발을 지원하고 있으며 **Apache 2.0 라이선스**로 무료 사용 가능

- EasyOCR : 이름 그대로 "가장 쉽고 직관적으로 사용할 수 있도록 만든 OCR(광학 문자 인식) 라이브러리"입니다.  
       - 네이버 산하의 AI 연구소인 Clova AI에서 공개한 딥러닝 모델들을 기반으로 개발되었으며, 복잡한 설명 없이 몇 줄의 코드로 80개 이상의 언어를 인식할 수 있어 현재 파이썬 생태계에서 매우 인기있는 도구입니다.

- PaddieOCR: 한글 및 표 인식률이 매우 뛰어난 라이브러리

- Cloud API: Google Vision, Naver CLOVA OCR 등

```python
# EasyOCR 라이브러리를 불러옵니다.
# 이미지에서 글자를 인식(OCR)하는 딥러닝 기반 라이브러리입니다.
import easyocr

# PyMuPDF 라이브러리를 불러옵니다.
# PDF 파일을 열고 페이지를 이미지로 변환하는 데 사용합니다.
import fitz

# OCR 리더기 초기화
# ['ko', 'en'] : 한국어와 영어를 인식하도록 설정합니다.
# 처음 실행하면 OCR 모델 파일을 자동으로 다운로드하므로 약 1~2분 정도 시간이 걸릴 수 있습니다.
reader = easyocr.Reader(['ko', 'en'])


# PDF 파일을 엽니다.
with fitz.open("data/table_example.pdf") as doc:

    # PDF의 두 번째 페이지를 선택합니다.
    # PyMuPDF는 페이지 번호가 0부터 시작합니다.
    page = doc[1]


    # page.get_pixmap() : PDF 페이지를 이미지로 변환합니다.
    # Matrix(3,3)는 해상도를 3배로 확대하여 OCR 정확도를 높이기 위한 설정입니다.
    pix = page.get_pixmap(matrix=fitz.Matrix(3, 3)) # (2, 2)로 테스트한 결과와 비교해 보세요.


    # 변환된 이미지를 임시 파일로 저장할 경로입니다.
    img_path = "output/temp_page.png"


    # PDF 페이지 이미지를 PNG 파일로 저장합니다.
    pix.save(img_path)

    # EasyOCR을 사용하여 이미지에서 텍스트를 추출합니다.
    # readtext() : 이미지에서 글자를 찾아 텍스트로 변환합니다.
    # detail=0 : 좌표 정보 등은 제외하고 텍스트만 리스트 형태로 반환합니다.
    result = reader.readtext(img_path, detail=0)


    # OCR 결과 출력
    print(f"--- [페이지 추출 결과] ---")


    # result는 텍스트 리스트 형태이므로 for문을 이용해 하나씩 출력합니다.
    for text in result:
        print(text)
```

```python
from img2table.document import PDF
from img2table.ocr import EasyOCR
import pandas as pd

# 1. EasyOCR 엔진 설정
# 한글('ko')과 영어('en')를 인식하도록 설정합니다.
ocr = EasyOCR(lang=["ko", "en"])

# 2. PDF 파일 경로 지정
pdf_path = "data/table_example.pdf"
pdf = PDF(src=pdf_path, pages=[3]) # 4번째 페이지(인덱스 3)만 지정

# 3. 표 추출 실행
# ocr 엔진을 전달하면, 표 영역을 찾은 뒤 그 안의 글자를 읽습니다.
# implicit_rows=True로 설정하면 눈에 보이는 선이 없어도 표 구조를 최대한 찾아냅니다.
extracted_tables = pdf.extract_tables(
    ocr=ocr,
    implicit_rows=True,     # 선이 불분명해도 행 구조를 유추함
    borderless_tables=True, # 테두리 선이 없는 표까지 검색 범위 확대
    min_confidence=30        # 표로 인식하는 기준치를 낮춰 더 많은 데이터를 잡음 (기본값 보통 50)
)

# 4. 결과 출력 및 데이터 프레임 변환
print(f"--- [표 추출 결과] ---")

# 추출된 표들을 순회하며 데이터 프레임으로 출력합니다.
for page_num, tables in extracted_tables.items():
    for i, table in enumerate(tables):
        print(f"\n[페이지 {page_num+1}의 {i+1}번쨰 표]")
        df = table.df  # 표를 Pandas DataFrame으로 변환

        # 결과 확인
        display(df) # 주피터 노트북 환경인 경우
        # print(df) # 터미널 환경인 경우

        df.to_excel(f"output/page_{page_num+1}_table_{i+1}.xlsx", index=False)
```
