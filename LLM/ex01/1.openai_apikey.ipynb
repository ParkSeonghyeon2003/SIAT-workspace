### load_dotenv:
- 프로젝트 루트의 .env 파일을 읽어 환경변수로 등록합니다.
- API Key(OpenAI Key 등)를 코드에 직접 적지 않고 안전하게 외부 파일(.env)에서 불러올 때 사용합니다.
- 환경 변수 설정은 LLM 호출의 필수 준비 단계입니다.

````python
%pip install openai python-dotenv
````

````python
# 운영체제의 환경변수(OS 환경변수)를 가져오거나 확인할 때 사용합니다.
import os

from dotenv import load_dotenv

# .env 파일 로드
load_dotenv()

OPENAI_KEY = os.getenv("OPENAI_API_KEY")

if OPENAI_KEY is None:
    raise EnvironmentError(
        "환경변수 OPENAI_API_KEY가 설정되어 있지 않습니다.\n"
        "1) 프로젝트 루트에 .env 파일을 만들고 아래처럼 넣으세요.\n"
        "   OPENAI_API_KEY=sk-여기에-실제-키\n"
        "2) 또는 시스템 환경변수에 OPENAI_API_KEY를 설정하세요\n"
        "3) 환경변수 설정 후 파이썬을 다시 실행하세요."
    )

print("OPENAI_API_KEY가 로드되었습니다. (길이:", len(OPENAI_KEY), "문자)")
````

### **주요 특징 및 장점**

- **상태 관리 (Stateful)**: 이전 API와 달리, 매 요청마다 전체 대화 기록을 보낼 필요가 없음 → 대화의 상태를 나타내는 ID를 전달해, OpenAI 서버 측에서 대화의 맥락을 관리 → 성능 향상 및 비용 절감

- **내장 도구 (Built-in Tools) 지원**: 웹 검색, 파일 검색, 컴퓨터 사용, 코드 인터프리터 등 강력한 도구들을 API 수준에서 기본으로 제공

- **단일화된 인터페이스**: Chat Completions API의 단순성과 Assistant API의 도구 사용 및 상태 관리 기능을 하나의 인터페이스로 결합 → 개발 복잡성 감소

- **향상된 추론 능력**: 특히 o1, o3, o4-mini 같은 최신 추론 모델과 함께 사용할 때, 이전 응답의 추론 과정을 활용해 더 높은 인텔리전스의 정확도를 발휘

- **멀티모달 지원**: 텍스트 및 이미지 입력을 모두 지원 → 더 다양한 애플리케이션 구축 가능

    - 모달은 정보의 형태 또는 데이터의 종류를 의미합니다.

        | 모달 | 예 |
        | :--- | :--- |
        | 텍스트 | 문장, 문서 |
        | 이미지 | 사진, 그림 |
        | 음성 | 사람의 말 |
        | 영상 | 동영상 |
        | 표 | 테이블 데이터 |

    - 멀티모달(Multimodal)이란 AI가 여러 가지 형태의 데이터(모달, Modal) 를 함께 이해하고 처리하는 기술을 의미합니다. 즉, 텍스트만 처리하는 것이 아니라 여러 유형의 정보를 동시에 다루는 AI입니다.

    - 다양한 입력 값 처리
