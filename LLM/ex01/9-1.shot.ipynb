Shot은 AI에게 보여주는 예제(example) 한 개를 의미합니다.

Shot의 구조

- Shot은 보통 질문 + 답변 한 쌍으로 이루어집니다.

    EXAMPLE_SHOTS = [ {"role": "user", "content": "참새"}, {"role": "assistant", "content": "짹짹!!!"}, ]

    이 부분이 1 shot입니다.

- {"role": "user", "content": "참새"} 의미

    | 항목 | 의미 |
    | :--- | :--- |
    | role=user | 사용자가 질문 |
    | content=참새 | 질문 내용 |

- {"role": "assistant", "content": "짹짹!!!"}

    | 항목 | 의미 |
    | :--- | :--- |
    | role=assistant | AI 답변 |
    | content=짹짹!!! | 답변 내용 |

    위 코드는 AI에게 다음 패턴을 보여주는 것입니다.

    참새 → 짹짹!!!

    AI는 이 예시를 보고 강아지 → 멍멍!!! 처럼 비슷한 방식으로 답하게 됩니다.

 ---
Shot 개수에 따른 종류

| Shot 종류 | 의미 |
| :--- | :--- |
| Zero-shot | 예제 없이 질문 |
| One-shot | 예제 1개 제공 |
| Few-shot | 예제 여러 개 제공 |
