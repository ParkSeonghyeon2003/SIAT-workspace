### **Conversations API**

https://platform.openai.com/docs/api-reference/conversations

- 대화형 AI 애플리케이션을 위한 **상태 관리(stateful) 기능을 제공**하는 최산 API 인터페이스
- 개발자가 매번 전체 대화 기록을 전송할 필요 없이 효율적으로 대화의 맥락을 유지하고 관리할 수 있도록 설계
- 서버 측에서 대화의 상태를 ID(conversation_id)로 관리  
→ 개발자는 이 ID를 사용해 메시지를 추가하거나 검색할 수 있으며, 모델은 ID에 연결된 전체 대화 기록을 참조해 응답을 생성
    - dash 보드에서 확인할 수 있어요.
    - API 서버에 저장되어 있어요
    - https://platform.openai.com/logs?api=responses
- 매 요청마다 전체 대화 기록을 전송할 필요가 없으므로, 긴 대화에서 데이터 전송량과 처리 시간을 줄여 성능을 향상시키고 비용을 절감할 수 있음
