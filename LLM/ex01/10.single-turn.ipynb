## 싱글턴(Single-turn)과 멀티턴(Multi-turn) 대화 방식

### 1. 싱글턴 (Single-turn)

- 한 번의 질문 → 한 번의 답변

- 즉 대화 맥락(Context)이 없는 방식입니다.

    사용자 → 질문  
    AI → 답변 (대화 종료)

    | 특징 | 설명 |
    | :--- | :--- |
    | 대화 1번 | 질문 1개만 처리 |
    | 맥락 없음 | 이전 대화 기록을 사용하지 않음 |
    | 구조 단순 | 구현이 쉬움 |
    | 비용 절약 | 토큰 사용 적음 |

- 예
```python
    response = client.responses.create(
        model="gpt-4.1-nano",
        instructions="당신은 상담사입니다.",
        input="스트레스 줄이는 방법 알려줘"
    )
```

 ---
### 2. 멀티턴 (Multi-turn)

- 여러 번 질문하고 이어서 대화하는 방식

- 이전 대화를 함께 전달하여 대화 맥락(Context)을 유지하는 방식입니다.

    사용자 → 질문  
    AI → 답변  
    사용자 → 추가 질문  
    AI → 답변

    | 특징 | 설명 |
    | :--- | :--- |
    | 여러 번 대화 | 지속적인 대화 가능 |
    | 맥락 유지 | 이전 대화 기록을 함께 전달 |
    | 자연스러운 대화 | 챗봇 형태의 대화 가능 |
    | 토큰 사용 증가 | 이전 대화 기록을 계속 포함하기 때문 |

- 예)
```python
    messages = [
        {"role": "user", "content": "스트레스 줄이는 방법 알려줘"},
        {"role": "assistant", "content": "운동과 명상이 좋습니다"},
        {"role": "user", "content": "운동 중에서 추천해줘"}
    ]
```

- 대화 흐름:

    사용자 : 스트레스 줄이는 방법 알려줘  
    AI : 운동과 명상이 좋습니다.

    사용자 : 운동 중에서 추천해줘  
    AI : 걷기나 요가를 추천합니다.
