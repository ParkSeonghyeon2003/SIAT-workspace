Response API에서 **instructions**와 **input**은 역할이 다릅니다.

쉽게 말하면 AI의 역할 설정 vs 실제 질문의 차이입니다.

### 1. instructions

- 모델의 행동 방식(역할, 규칙, 스타일)을 설정하는 부분입니다.

- 예전 Chat Completions API의 system 메시지 역할과 같습니다.

- 예시

    instructions="You are a helpful assistant."

- 의미

    AI의 성격 설정

    답변 방식 지정

    말투, 규칙, 역할 지정

    예)

    instructions="당신은 한국 문학 교사입니다."

    이렇게 하면 모델은 문학 교사처럼 설명합니다.

    또는

    instructions="모든 답변을 한 문장으로 작성하세요."

    → 답변 형식을 제한할 수도 있습니다.

### 2. input

- 실제로 모델에게 물어보는 질문입니다.

- 예전 Chat Completions API의 user 메시지 역할입니다.

- 예)

    input="홍길동전을 한 문장으로 요약해줘"

    즉 AI에게 하는 실제 요청입니다.

### 3. assistant

- AI가 생성한 답변을 의미합니다.

- 예전 Chat Completions API의 assistant 메시지 역할입니다.

### 4. 구조 비교

| 구분 | 역할 | 예전 API |
| :--- | :--- | :--- |
| **instructions** | AI의 역할, 규칙 설정 | system |
| **input** | 실제 질문 | user |
| **assistant** | AI의 답변 | assitant |
