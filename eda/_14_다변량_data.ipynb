- 다중 변수 또는 다변량 데이터는 변수가 2개 이상인 데이터를 의미합니다.
- 다중 변수 데이터에 대한 분석은 변수와 변수 사이의 관계를 파악하는데 초점이 있습니다.
---
산점도(Scatter Plot)란 두 개의 연속형 변수 사이의 관계를 점으로 표현한 그래프입니다.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/mtcars.csv')

df.plot.scatter(x='wt', y='mpg')
plt.show()
```

- 변수 유형: wt(중량)와 mpg(연비)는 모두 연속적인 숫자로 이루어진 수치형(Numerical) 데이터입니다.
- 분석 목적: 두 변수를 각각 축으로 사용하는 산점도(Scatter Plot)를 통해, "차 무게가 무거워질수록 연비가 낮아지는가?"와 같은 두 변수 간의 상관관계를 파악하는 것이 목적입니다.
- 산점도를 보면 오른쪽 아래로 향하는 경향(음의 상관관계)을 확인할 수 있습니다. 즉, 차가 무거울수록 연비가 떨어지는 일반적인 상식과 일치하는 데이터임을 알 수 있습니다.
---

```python
df.plot.scatter(x='wt',
                y='mpg',
                s=10,
                c='red',
                marker='s') # o: 원, s; 사각형, ^: 삼각형, x: 엑스, D: 다이아
plt.show()
```

```python
vars = ['mpg', 'disp', 'hp', 'wt']

pd.plotting.scatter_matrix(df[vars])
plt.show()
```

- 다중 산점도(Scatter Matrix)는 여러 수치형 변수 간의 관계를 한눈에 파악하기 위해 사용합니다.
- 다중 산점도 읽는 법

    (1) 대각선 (Diagonal)
    - 자기 자신과의 관계를 나타내는 칸으로, 산점도 대신 해당 변수의 히스토그램(Histogram)이 그려집니다.
    - 이를 통해 각 변수의 데이터 분포(단일 변수 탐색) 를 확인할 수 있습니다.

    (2) 비대각선 (Off-diagonal): 두 변수 사이의 상관관계를 보여주는 산점도입니다.
    - mpg(연비)는 다른 모든 변수(disp, hp, wt)와 오른쪽 아래 방향을 향하고 있습니다. (강한 음의 상관관계)
    - 중량(wt)이 늘어남에 따라 배기량(disp)과 마력(hp)은 정비례 증가하고 있습니다.

```python
df = pd.read_csv('data/iris.csv')

color_map = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
colors = list(color_map[key] for key in df.Species)
print(colors)

df.plot.scatter(x='Petal_Length',
                y='Petal_Width',
                s=30,
                c=colors,
                marker='o')
plt.show()
```

- 종(Species)별 군집화(Clustering) 현상
    - 가장 눈에 띄는 분석 결과는 세 가지 색상의 점들이 특정 영역에 뭉쳐있다는 점입니다.
    - Setosa (빨간색)은 다른 종들과 완전히 분리되어 왼쪽 하단에 작게 뭉쳐있습니다.
    - Virginica (파란색)가 다른 두 품종에 비해 꽃잎의 길이와 폭이 큰 것을 알 수 있습니다.
    - 변수 간의 강한 양(+)의 상관관계 분석: 꽃잎의 길이가 길어질수록 너비도 함께 넓어지는 매우 밀접한 관계가 있습니다.

```python
df = pd.read_csv('data/iris.csv')

color_map = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
fig, ax = plt.subplots()

for species, group in df.groupby('Species'):
    ax.scatter(
        group['Petal_Length'],
        group['Petal_Width'],
        s=30,
        c=color_map[species],
        label=species
    )

ax.set_xlabel('Petal_Length')
ax.set_ylabel('Petal_Width')
ax.legend(title='Species')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

beers = [5, 2, 9, 8, 3, 7, 3, 5, 3, 5]
bal = [0.1, 0.03, 0.19, 0.12, 0.04, 0.0095, 0.07,
       0.06, 0.02, 0.05]

dict = {'beers': beers, 'bal': bal}
df = pd.DataFrame(dict)

df.plot.scatter(x='beers', y='bal',
                title='Beers-Blood Alcohol Level')
plt.show()
```
