특정 지역 지도 보기

```python
import folium
import geokakao as gk

import webbrowser
import os

def showMap(map):
    map.save('map.html')
    filepath = os.getcwd()
    file_uri = 'file:///' + filepath + '/map.html'
    webbrowser.open_new_tab(file_uri)
```

```python
loc = [37.54, 127.05]
map = folium.Map(location=loc)
showMap(map)
```

마커 모양 변경하기

```python
import folium
import geokakao as gk

loc = gk.convert_address_to_coordinates('강원특별자치도 강릉시 창해로 514')
loc
```

```python
map = folium.Map(location=loc, zoom_start=13)
folium.Marker(location=loc,
            icon=folium.Icon(color='red', icon='star', prefix='fa')).add_to(map)
showMap(map)
```

icon='star' → 사용할 아이콘의 이름입니다.

prefix='fa' → 이 아이콘이 Font Awesom(Font Awesome Icons) 라이브러리 소속임을 의미합니다.

prefix를 사용하는 아이콘은 여러 라이브러리 중 하나를 선택해서 가져올 수 있습니다.

마커 위치에 텍스트 추가하기

```python
import folium
import geokakao as gk

loc = gk.convert_address_to_coordinates('강원특별자치도 강릉시 창해로 514')
print(loc)

map = folium.Map(location=loc, zoom_start=13)
folium.Marker(location=loc,
              icon=folium.Icon(color='red', icon='star')).add_to(map)
showMap(map)
```

```python
html_start = '<div \
style="\
font-size: 12px; \
color: blue; \
background-color: rgba(255, 255, 255, 0.2); \
width: 85px; \
text-align: left; \
margin: 0px;\
"><b>'
html_end = '</b></div>'
```

```python
folium.Marker(location=loc,
            icon=folium.DivIcon(
                icon_anchor=(0, 0), # 텍스트 위치 설정
                html=html_start+'경포해수욕장'+html_end
            )).add_to(map)

showMap(map)
```

다수의 마커 표시하기

```python
import folium
import geokakao as gk
import pandas as pd
# 관광지 정보를 데이터프레임으로 저장
names = ['용두암', '성산일출봉', '정방폭포',
         '중문관광단지', '한라산1100고지', '차귀도']
addr = ['제주시 용두암길 15',
        '서귀포시 성산읍 성산리',
        '서귀포시 동홍동 299-3',
        '서귀포시 중문동 2624-1',
        '서귀포시 색달동 산1-2',
        '제주시 한경면 고산리 산 117']
dict = {"names": names, "addr": addr}
df = pd.DataFrame(dict)
print(df)
```

```python
gk.add_coordinates_to_dataframe(df, 'addr')
print(df)
print(df.dtypes)
```

```python
df.decimalLatitude = pd.to_numeric(df.decimalLatitude)
df.decimalLongitude = pd.to_numeric(df.decimalLongitude)
df.dtypes
```

```python
center = [df.decimalLatitude.mean(), df.decimalLongitude.mean()]
center
```

```python
map = folium.Map(location=center, zoom_start=10)

for _, row in df.iterrows():
    folium.Marker(location=[row.iloc[2], row.iloc[3]],
                icon=folium.Icon(color='red', icon='star')).add_to(map)
showMap(map)
```

```python
html_start = '<div \
style="\
font-size: 12px; \
color: blue; \
background-color: rgba(255, 255, 255, 0.2); \
width: 85px; \
text-align: left; \
margin: 0px;\
"><b>'
html_end = '</b></div>'
```

```python
for _, row in df.iterrows():
    folium.Marker(location=[row.iloc[2], row.iloc[3]],
                icon=folium.DivIcon(
                    icon_anchor=(0, 0), # 텍스트 위치 설정
                    html=html_start+row['names']+html_end
                )).add_to(map)

showMap(map)
```
