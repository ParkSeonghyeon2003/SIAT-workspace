---
### 2차원 리스트를 데이터프레임으로 변환

```python
score=[['정약용', 70, 80, 100],
       ['이순신', 100, 80, 70],
       ['이율곡', 90, 95, 90]]
print(type(score))
print(score)
```

---
DataFrame으로 바꾸기

```python
import pandas as pd
score_df = pd.DataFrame(score)

print(type(score_df))
print(score_df)
```

---
아래와 같이 출력되도록 합니다  

```python
score_df.columns = ["이름", "국어", "영어", "수학"]
print(score_df)
```

---
아래와 같이 출력되도록 합니다   

```python
score_df.info()
```

```python
# 원본 복사
score_df_copy = score_df.copy()
```

---
이름이 이순신 인지 비교

```python
print(score_df_copy["이름"] == '이순신')
```

---
이름이 이순신인 행 출력

```python
print(score_df_copy[score_df_copy["이름"] == "이순신"])
```

---
이름이 이순신인 행의 index 구하기

```python
idx = score_df_copy[score_df_copy["이름"] == "이순신"].index
print(idx)
```

---
이름이 이순신인 행 삭제하기

```python
result = score_df_copy.drop(index=idx)
```

```python
print(result)
```

---
수학 컬럼 삭제하기

```python
score_df_copy = score_df_copy.drop(columns=["수학"])
print(score_df_copy)
```

---
과학 컬럼을 국어 앞에 추가하기

```python
score_df_copy.insert(loc=1, column="과학", value=[80, 90, 86])
print(score_df_copy)
```
