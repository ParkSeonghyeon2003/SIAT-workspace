문제4. 사분위수를 기준으로 4개 구간의 집값 평균을 구하고 이를 막대 그래프로 시각화하세요.

- grp = pd.qcut(house_price, q=4, labels=['Q1', 'Q2', 'Q3', 'Q4']) : 데이터를 개수가 거의 동일하도록 자동으로 4개 구간(사분위수) 으로 분할

- house_price.groupby(grp).mean() : 각 사분위(Q1~Q4)에 속한 주택 가격의 평균값 계산

```python
grp = pd.qcut(house_price, q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
avg = house_price.groupby(grp).mean()

avg.plot
plt.xlabel(         )
plt.ylabel(          )
plt.title(             )
plt.show()
```
