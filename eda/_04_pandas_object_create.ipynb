## 판다스
- 판다스(Pandas = Panel Data + Python Data Analysis)는 **파이썬(Python)에서 데이터를 다루기 위한 라이브러리**로,
    특히 표(테이블) 형태의 데이터(행과 열) 를 쉽게 처리할 수 있도록 도와줍니다.

### 주요 특징
1. **데이터 구조 제공**
    - **Series**: 1차원 데이터(리스트와 비슷하지만 인덱스 포함)
    - **DataFrame**: 2차원 데이터(엑셀 표처럼 행과 열로 구성)
2. **데이터 읽기/쓰기**
    - CSV, Excel, SQL, JSON 등 다양한 파일을 읽고 저장 가능
3. **데이터 가공**
    - 결측치 처리, 중복 제거, 정렬, 필터링, 그룹화 등 가능
4. **통계 및 집계**
    - 평균, 합계, 최대/최소, 표준편차같은 기본 통계 연산 지원
5. **다른 라이브러리와 연계**
    - NumPy, Matplotlib, Scikit-learn 같은 라이브러리와 함께 자주 사용

---
2차원 리스트를 데이터프레임으로 변환

```python
score = [["정약용", 70, 80, 100],
         ["이순신", 100, 80, 70],
         ["이율곡", 90, 95, 90]]
print(type(score))
print(score)
```

```python
score_df = pd.DataFrame(score)
print(type(score_df))
```

```python
print(score_df)
```

```python
score_df.columns = ["이름", "국어", "영어", "수학"]
print(score_df)
```

```python
score_df.info()
```

```python
score_df.index = ["1번", "2번", "3번"]
print(score_df)
```

---
딕셔너리를 DataFrame으로 변환

```python
n = pd.DataFrame({
    "col_str1": ["1", "2", "3", "4"],
    "col_str2": ["5", "6", "7", "8"]
})

print(n)
```

```python
print(n.dtypes) # 각 컬럼은 문자열 자료형을 저장하므로 object다
```

```python
n_df = n.astype({"col_str1":int, "col_str2":float}) # 타입 변경
print(n_df)
```

```python
print(n_df.dtypes)
```
