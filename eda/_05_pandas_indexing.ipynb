```python
import pandas as pd
score = [78, 50, 90, 60, 100]
```

### 판다스 시리즈(Series) 객체 생성

```python
nums = pd.Series(score)
print(nums)
```

```python
nums.index = ["python", "HTML", "CSS", "javascript", "react"]
print(nums)
```

---
판다스의 인덱싱(원하는 위치나 이름을 지정해서 값을 선택하는 것)
1. 값의 절대 위치에 의한 인덱싱
    - iloc[] 사용
    - 예) nums.iloc[2]
2. 값의 레이블에 의한 인덱싱
    - loc[] 사용
    - 예) nums.loc["CSS"]

```python
print(nums.iloc[2])
```

```python
print(nums.loc["CSS"])
```

### 판다스 DataFrame 객체 생성

```python
score = [
    [85, 72, 90, 68, 95],
    [78, 88, 65, 80, 92],
    [90, 76, 88, 85, 70]
]
```

```python
df = pd.DataFrame(score)
print(df)
```

```python
df.columns = ["python", "HTML", "CSS", "javascript", "react"]
df.index = ["Lee", "Kim", "Park"]
print(df)
```

```python
print(df.iloc[1, 2])
print(df.loc["Kim", "CSS"])
```

---
### 중복된 레이블
- 절대 위치 인덱스는 특성상 중복이 존재하지 않고 시스템에 의해 관리됩니다.
- 레이블 인덱스는 사용자가 임의로 지정할 수 있고 중복이 가능합니다.

```python
scores = pd.Series(
    [85, 90, 78, 88, 92],
    index=["철수", "영희", "철수", "민수", "영희"]
)

print(scores)
```

```python
print(scores["철수"])
```

### 숫자 레이블 부여하기

```python
s = pd.Series(
    [80, 90, 75, 88],
    index=[101, 102, 103, 104]
)

print(s)
```

```python
print(s.loc[102])
```

```python
print(s.iloc[1])
```
