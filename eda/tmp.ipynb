문제 18. 부서별 사원 수와 평균 급여를 한 번에 구하세요.

- .groupby('DEPTNO'):데이터프레임을 'DEPTNO'(부서번호) 컬럼의 값을 기준으로 끼리끼리 묶습니다.
- .agg(...): 묶인 그룹별로 어떤 계산을 할지 정의합니다.
- emp_count=('EMPNO', 'count'):
    - 새로 만들어질 컬럼 이름을 **emp_count**라고 정합니다.
    - EMPNO(사원번호) 컬럼을 선택하여 그 개수(count)를 셉니다.
- avg_sal=('SAL', 'mean'):
   -  새로 만들어질 컬럼 이름을 **avg_sal**이라고 정합니다.
   -  SAL(급여) 컬럼을 선택하여 평균(mean)을 구합니다. (즉, 부서별 평균 급여) 

```python
result = emp.groupby('DEPTNO').agg(
    emp_count=('EMPNO', 'count'),
    avg_sal=('SAL', 'mean')
)
```

```python
print(result)
```
