# 인공 신경망 (Artificial Neural Network, ANN)

- 생물학적 뉴런(Neuron)의 구조와 작동 방식을 모방하여 만든 머신러닝 알고리즘입니다.

- 여러 개의 인공 뉴런을 연결하여 데이터를 학습하고 예측을 수행합니다.

- 인공 신경망은 이미지, 음성, 텍스트처럼 복잡한 패턴을 가진 데이터를 처리하는 데 매우 뛰어난 성능을 보입니다.

- 특히 **신경망의 층(layer)을 깊게 쌓은 구조를 딥러닝(Deep Learning)** 이라고 합니다.

    - **사용자 친화성**

    - **빠른 실험**

## 2. 텐서플로우 (TensorFlow)

- 구글(Google)에서 개발한 오픈소스 딥러닝 프레임워크로  
현재 전 세계에서 가장 널리 사용되는 딥러닝 개발 도구 중 하나입니다.

- 역할  
신경망을 구성하는 **대규모 수치 연산을 효율적으로처리하는 엔진**입니다.  
이름의 의미는 다음과 같습니다.
    - **Tensor** : 다차원 배열(행렬, 이미지, 데이터 등)

- 특징
    - **저수준(Low-level) API** 제공 → 매우 세밀한 모델 설계 가능
    - **GPU / TPU 가속 지원**
    - **모바일, 웹, 클라우드 배포 지원 (TensorFlow Lite, TensorFlow.js 등)
    - 연산 구조가 복잡하여 **초보자가 직접 사용하기에는 다소 어려움**

같은 코드와 같은 데이터로 동일한 결과를 얻기 위한 설정(재현성)

```python
import tensorflow as tf

# TensorFlow / NumPy / Python에서 사용하는
# 난수 생성 시드(seed)를 42로 고정합니다.
# 같은 시드를 사용하면 난수 생성의 시작 상태가 동일해집니다.
tf.keras.utils.set_random_seed(42)

# GPU는 병렬 연산을 수행하기 때문에
# 연산 순서가 실행 시점마다 달라질 수 있습니다.
# 이로 인해 같은 코드라도 결과가 미세하게 달라지는
# 비결정적(non-deterministic) 연산이 발생할 수 있습니다.
# 따라서 시드를 고정하더라도 100% 동일한 결과가
# 항상 보장되지는 않습니다.

# TensorFlow 연산에서 발생하는 비경정성을 최소화하기 위헤
# 가능한 한 연산 순서와 알고리즘을 고정하도록 설정합니다.
tf.config.experimental.enable_op_determinism()
```

## 로지스틱 회귀로 패션 아이템 분류하기

- 전체 데이터를 한꺼번에 사용하여 모델을 훈련하는 것보다 샘플을 하나씩 꺼내서 모델을 훈련하는 방법이 더 효율적

- 확률적 경사 하강법을 사용

- 패션 MNIST의 경우 각 픽셀은 0에서 255 사이의 정숫값을 가지며, 이런 이미지의 경우 보통 255로 나누어 0~1 사이의 값으로 정규화

- 이는 표준환느 아니지만 양수 값으로 이루어진 이미지를 전처리할 때 널리 사용하는 방법

```python
# 전처리
# 255로 나눠서 0~1 범위로 정규화(normalization)
train_scaled = train_input / 255.0

# (60000, 28, 28) → 60,000개의 28 x 28 이미지
# => (60000, 28 * 28) 로 변환
train_scaled = train_scaled.reshape(-1, 28 * 28)
```

```python
print(train_scaled.shape)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import SGDClassifier

# SGD 기반 로지스틱 회귀 모델 생성
# loss='log_loss': 로지스틱 회귀에서 사용하는 손실 함수
sc = SGDClassifier(loss='log_loss', max_iter=5, random_state=42)

# 5-fold cross validate 수행 (데이터를 5등분하여 각 fold에서 학습/검증)
scores = cross_validate(sc, train_scaled, train_target, n_jobs=-1)
print(scores)

# scores['test_scores'] → 검증용 폴드에서 계산된 정확도(accuracy)
# np.mean() → 5-fold 정확도의 평균값
print(np.mean(scores['test_score']))
```

# 로지스틱 회귀로 패션 아이템 분류하기

- 패션 MNIST 데이터(28x28=784 픽셀)에 맞게 로지스틱 회귀 방정식을 구성하면 다음과 같습니다.

    - 각 클래스(티셔츠, 바지 등)마다 별도의 선형 방정식이 만들어집니다.

        - $z_{티셔츠} = w1 \times (픽셀1) + w2 \times (픽셀2) + \dots + w784 \times (픽셀784) + b$

        - $z_{바지} = w1' \times (픽셀1) + w2' \times (픽셀2) + \dots + w784' \times (픽셀784) + b'$

- SGD Classifier 모델의 학습 원리

    - 패션 MNIST 데이터의 클래스를 가능한 잘 구분할 수 있도록 10개의 방정식에 대한 모델 파라미터(가중치 $w$와 절편 $b$)를 탐색합니다.



 ---
## 인공 신경망(ANN - Artificial Neural Network)으로의 연결

로지스틱 회귀의 이 구조는 인공 신경망의 가장 기본적인 층(Dense Layer)과 동일한 구조를 가집니다.

- 입력층 (Input Layer): 784개의 픽셀 값($x_1, x_2, \dots, x_784$)

- 가중치 (Weights): 각 연결선에 부여된 값 (이미지의 특징을 결정)

- 출력층 (Output Layer): 10개의 클래스별 $z$값 (가장 높은 $z$값이 최종 예측 결과가 됨)  
    -인공 신경망에서 z값을 계산하는 단위를 뉴런 또는 유닛이라고 합니다.

- 티셔츠를 계산할 때 쓰는 784개의 가중치는 바지를 계산할 때 쓰는 가중치와 완전히 별개입니다. 모델은 학습을 통해 각 옷의 특징을 가장 잘 나타내는 $w$값들을 찾아냅니다.

## 인공 신경망으로 모델 만들기

- 로지스틱 회귀에서는 교차 검증을 사용해 모델을 평가했습니다.
- 인공 신경망에서는 교차 검증을 잘 사용하지 않고 검증 세트를 별도로 덜어내어 사용합니다.
    - 딥러닝 분야의 데이터 셋은 충분히 크기 때문에 검증 점수가 안정적입니다.
    - 교차 검증을 수행하기에는 훈련 시간이 너무 오래 걸립니다.

```python
from sklearn.model_selection import train_test_split

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
print(train_scaled.shape, train_target.shape)
```

```python
print(val_scaled.shape, val_target.shape)
```

 ---
입력층 정의하기
- 입력층은 입력 그 자체이고 별다른 연산을 수행하지 않습니다.



```python
# keras.layers.input()는 입력층을 정의합니다.
inputs = keras.layers.Input(shape=(784,))
```

 ---
밀집층



- 가장 간단한 인공 신경망의 층입니다.

- 인공 신경망에는 여러 종류의 층이 있습니다.

- 밀집 층에서는 뉴런들이 모두 연결되어 있기 때문에 완전 연결 층이라고도 부릅니다.

- 특별히 출력층에 밀집층을 사용할 때는 분류하려는 클래스와 동일한 개수의 뉴런을 사용합니다.

 ---
### 소프트맥스와 같이 뉴런의 선형 방정식 계산 결과에 적용하는 함수를 **활성화 함수**라고 합니다.



```python
# 10개의 뉴련에서 출력되는 값을 확률로 바꾸기 위해서는 소프트맥스를 사용합니다.

# 10은 뉴런 개수로 10개의 패션 아이템을 분류하기 위해 사용
# activation='softmax' 뉴련 출력에 적용할 함수
dense = keras.layers.Dense(10, activation='softmax')
```

```python
# keras.Sequential : 층(layer)을 순차적으로 쌓는 방식의 모델
model = keras.Sequential([inputs, dense])
```

 ---
### 1. loss='sparse_categorical_crossentropy' 손실함수 지정한 경우

정답 레이블이 정수 형태일 때 사용하는 손실 함수입니다.

```python
print(train_target[:10])
```

```python
# 케라스 모델을 학습할 준비를 시키는 단계
# 즉, 모델이 학습할 때 필요한 손실 함수, 최적화 알고리즘, 평가 지표 등을 설정하는 과정입니다.
# 1. loss='sparse_categorical_crossentropy' 손실함수 지정
#    정답 레이블이 정수 형태일 때 사용하는 손실 함수입니다.
# 2. metrics=['accuracy']는 정확도를 추가로 계산해서 보여주라는 의미입니다.
#    케라스 모델이 훈련할 때 기본으로 에포크마다 손실 값을 출력합니다.
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# 모델 학습(훈련) 을 실행하는 함수힙니다.
# epoch=5 : 전체 훈련 데이터를 5번 반복 학습합니다.
# accuracy: 0.8538 - loss: 0.4295 : 정확도와 손실값을 출력합니다.
model.fit(train_scaled, train_target, epochs=5)
```

```python
model.evaluate(val_scaled, val_target)
```
