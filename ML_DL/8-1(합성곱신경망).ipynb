## 1. 합성곱 신경망(CNN, Convolutional Neural Network)

CNN은 **이미지·영상처럼 공간적 구조(가로·세로·채널)를 가진 데이터**를 잘 처리하도록 설계된 딥러닝 신경망입니다.  
사람이 이미지를 인식할 때 **부분 → 전체**로 이해하는 방식과 유사하게 동작합니다.

작은 필터(filter)를 이용해 이미지의 **국소적인 영역을 조금씩 훑으면서 특징을 추출**하고,  
이렇게 추출된 특징을 여려 층에 걸쳐 쌓아가며 점점 더 **고수준의 특징(high-level feature)** 을 학습합니다.

---

### 합성곱(Convolution)이란

합성곱은 **작은 필터(filter)를 입력 데이터 위에서 이동시키면서 계산하는 연산**입니다.  
각 위치에서 **입력 값과 필터 값을 곱하고 더하여 새로운 값을 생성**합니다.

---

### 커널과 필터

1. 커널 (Kernel)

- 커널은 2차원 행렬 하나를 의미합니다.
- 이미지의 가로($W$)와 세로($H$)에 대응하는 가중치 묶음입니다.
- 예를 들어, $3 \times 3$ 크기의 행렬 하나가 바로 커널입니다.
- 수학적 단위: 단일 채널(흑백 이미지 등)에서 특징을 추출하는 최소 단위입니다.

2. 필터 (Filter)

- 필터는 커널들의 집합이며, 입력 데이터의 채널(Channel) 개수를 포함하는 개념입니다.
- 입력 이미지가 RGB(3채널)라면, 필터 하나는 3개의 커널을 가집니다. (빨강용 커널, 초록용 커널, 파랑용 커널)
- 수학적 단위: 필터 1개가 연산을 마치면 결과물로 특성 맵(Feature Map) 1장이 나옵니다.

 ---
## 2. CNN의 기본 구성 요소

### ① 합성곱층 (Convolution Layer)

**역할**

- 이미지의 **국소적인 특징(feature)** 을 추출합니다.
- 예: 모서리, 선, 질감, 패턴

**동작 방식**

- 작은 **필터** 가 이미지 위를 슬라이딩하면서
- 각 위치에서 **곱셈 + 덧셈 연산**을 수행하여 **특성 맵(feature map)** 을 생성합니다.

**특징**

- 같은 필터를 전체 이미지에 공유 → 파라미터 수 감소
- 여러 개의 필터 → 여러 개의 특성 맵 생성
- 보통 ReLU 같은 **활성화 함수**와 함께 사용합니다.

**계층별 특징**

- 얕은 층: 모서리, 방향성 같은 저수준 특징
- 깊은 층: 눈, 코, 사물 형태 같은 고수준 특징

---

### ② 풀링층 (Pooling Layer)

**역할**

- 특성 맵의 **공간 크기를 줄이면서 중요한 정보만 유지**합니다.
- 계산량 감소 및 과적합 완화 효과가 있습니다.

**대표적인 방식**

- **Max Pooling**: 영역 내 최댓값 선택
- Average Pooling: 영역 내 평균값 선택

**특징**

- 학습할 파라미터가 없습니다.
- 위치가 조금 변해도 비슷한 특징으로 인식할 수 있습니다. → 위치 변화에 강건(translation invariance)

---

### ③ 밀집층 (Dense / Fully Connected Layer)

**역할**

- 앞에서 추출된 특징들을 **종합하여 최종 예측**을 수행합니다.
- 일반적인 신경망(MPL)과 동일한 구조입니다.

**동작**

- 합성곱/풀링 결과를 **1차원으로 펼친 후(flatten)**
- 모든 뉴런이 서로 연결된 구조로 분류 또는 회귀를 수행합니다.

**출력층 예**

- 이진 분류: sigmoid
- 다중 분류: softmax

---

## 3. CNN 전체 구조 흐름
```text
    입력 이미지
        ↓
    합성곱층 + 활성화 함수
        ↓
    풀링층
        ↓
    (합성곱 + 풀링 반복)
        ↓
    Flatten
        ↓
    밀집층
        ↓
    출력 (분류 / 예측)
```

 ---
## 4. 구성 요소 정리 표

| 구성 요소 | 역할 | 학습 파라미터 |
| :--- | :--- | :--- |
| 합성곱층 | 국소 특징 추출 | 있음 |
| 풀링층 | 크기 축소, 핵심 정보 유지 | 없음 |
| 밀집층 | 최종 판단 | 있음 |

---

## 5. 핵심 정리

- CNN은 **이미지의 공간 구조를 유지한 채 특징을 학습**합니다.
- 합성곱/풀링층은 **특징 추출기**
- 밀집층은 **의사결정기** 역할을 합니다.
